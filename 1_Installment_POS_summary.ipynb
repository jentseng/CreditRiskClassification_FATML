## Pay Summary

```python
def greater(series):
    if series >= 1:
        return 1
    else:
        return 0
```

```python
df_pay_filled['DPD'] = df_pay_filled['DAYS_ENTRY_PAYMENT']-df_pay_filled['DAYS_INSTALMENT']
df_pay_filled['DPD_Count'] = df_pay_filled['DPD'].map(greater)
#if PAID/NOT = 1, then they have not paid that loan, 0 means they have made a payment
df_pay_filled['NOT_PAID'] = df_pay_filled['DAYS_ENTRY_PAYMENT'].map(greater)
df_pay_filled['UNDER_PAY'] = df_pay_filled['AMT_INSTALMENT']-df_pay_filled['AMT_PAYMENT']

mask1 = df_pay_filled['DAYS_ENTRY_PAYMENT'] != 1
mask2 = df_pay_filled['AMT_INSTALMENT'] > df_pay_filled['AMT_PAYMENT']

df_pay_filled['UNDER_PAY_COUNT']=df_pay_filled[mask1&mask2]['UNDER_PAY'] > 0
```

```python
df_pay_filled.sort_values(by='SK_ID_CURR')[:10]
```

```python
df_pay_filled.sort_values(by='UNDER_PAY_COUNT',ascending=False)[:10]
```

```python
df_pay_filled[df_pay_filled['NOT_PAID']==1]['UNDER_PAY_COUNT'].sum()
```

```python
df_pay_filled['DPD_min'] = df_pay_filled['DPD']
df_pay_filled['DPD_avg'] = df_pay_filled['DPD']
```

```python
df_pay_filled['UNDER_PAY_COUNT'].sum()
```

```python
df_pay_group = df_pay_filled.groupby('SK_ID_CURR').agg({'NUM_INSTALMENT_NUMBER':'count','NOT_PAID':'sum', 'DPD':'max','DPD_min':'min','DPD_avg':'mean', 
                                                        'DPD_Count':'sum',
                                                        'AMT_INSTALMENT':'sum', 'AMT_PAYMENT':'sum', 'UNDER_PAY_COUNT':'sum'})
```

```python
#some could have underpaid on some and overpaid on others, how should we represent this?
df_pay_group.sort_values(by='NOT_PAID',ascending=False)[:10]
```

```python
df_pay_group['PERCENT_NOT_PAID'] = df_pay_group['NOT_PAID']/df_pay_group['NUM_INSTALMENT_NUMBER']
df_pay_group['OWED'] = df_pay_group['AMT_INSTALMENT']-df_pay_group['AMT_PAYMENT']
df_pay_group['PAYMENTS_MADE']=df_pay_group['NUM_INSTALMENT_NUMBER']-df_pay_group['NOT_PAID']

try:
    df_pay_group['PERCENT_UNDER_PAID'] = df_pay_group['UNDER_PAY_COUNT']/df_pay_group['NUM_INSTALMENT_NUMBER']
except ZeroDivisionError:
    print ("0")

df_pay_group['PERCENT_OWED'] = df_pay_group['OWED']/df_pay_group['AMT_INSTALMENT']
df_pay_group['no_DPD_Percentage'] = 1-(df_pay_group['DPD_Count']/df_pay_group['NUM_INSTALMENT_NUMBER'])
```

```python
#print(df_pay_group['DPD_Count'].value_counts()[:10])
df_pay_group.sort_values(by='PERCENT_OWED')
```
